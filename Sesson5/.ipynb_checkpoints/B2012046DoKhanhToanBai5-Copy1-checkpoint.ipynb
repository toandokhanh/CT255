Bài 1 Thực hiện lại các hướng dẫn từ trang 89 đến 90.


```python
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
print(X.shape)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
	X, y, test_size=0.2, random_state=1)
from sklearn import svm
from sklearn.model_selection import GridSearchCV

tuned_parameters = [{'kernel' : ['rbf'], 'C' : [1,10,100]},
                   {'kernel' : ['linear'], 'C' : [1,10,100]}]
svm_reg = GridSearchCV (svm.SVR(), tuned_parameters, scoring = 'neg_mean_absolute_error', cv=2
                       )
svm_reg.fit(X_train, y_train)
svm_reg.best_params_
```

```python
from sklearn import svm
svm_regr = svm.SVR(kernel='linear', C=1)
svm_regr.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error, mean_absolute_error
y_train_pred = svm_regr.predict(X_train)
print('Train: Mean squared error = ', mean_squared_error(y_train,y_train_pred))
print('Train: Mean absolute error = ', mean_absolute_error(y_train,y_train_pred))

y_test_pred = svm_regr.predict(X_test)
print('Test: Mean squared error = ', mean_squared_error(y_test,y_test_pred))
print('Test: Mean absolute error = ', mean_absolute_error(y_test,y_test_pred))
```

Câu 3: Trong phần code lựa chọn siêu tham số ở [11] hình 4.31 trang 90 với 2 siêu tham số là kernel và C, bạn hãy tự đề xuất những giá trị khác của bộ siêu tham số này (Vd: C=[50, 100], kernel = ‘poly’, ‘rbf’, ‘sigmoid’,....) để đạt được độ lỗi tốt hơn trong giáo trình (Hiện giá trị MAE bộ test trong giáo trình là 3.48). Cho biết giá trị bộ siêu tham số đạt kết quả tốt nhất là gì (vd: trong giáo trình là C=1, kernel=linear sẽ đạt kết quả tốt nhất).


```python
from sklearn.datasets import load_boston
X, y = load_boston(return_X_y=True)
print(X.shape)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
	X, y, test_size=0.2, random_state=1)

from sklearn import svm
from sklearn.model_selection import GridSearchCV

tuned_parameters = [{'kernel': ['poly'], 'C': [50, 100]},
                   {'kernel': ['sigmoid'], 'C': [50, 100]},
                   {'kernel': ['rbf'], 'C': [50, 100]}]

svm_reg = GridSearchCV(
    	svm.SVR(), tuned_parameters,
    	scoring = 'neg_mean_absolute_error', cv=2
	)
svm_reg.fit(X_train, y_train)
svm_reg.best_params_
```

```python
from sklearn import svm
svm_regr = svm.SVR(kernel='linear', C=2)
svm_regr.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error, mean_absolute_error
y_train_pred = svm_regr.predict(X_train)
print('Train: Mean squared error = ',mean_squared_error(y_train, y_train_pred))
print('Train: Mean absolute error = ',mean_absolute_error(y_train, y_train_pred))

y_test_pred = svm_regr.predict(X_test)
print('Test: Mean squared error = ',mean_squared_error(y_test, y_test_pred))
print('Test: Mean absolute error = ',mean_absolute_error(y_test, y_test_pred))
```

câu 4: Sử dụng khối [1] ở hình 4.13 trang 78 để load dữ liệu digits, kết hợp với [11] ở hình 4.31 để lựa chọn siêu tham số cho bài toán phân lớp 10 số ở dataset digits. Gợi ý: Dòng 8 khối lệnh 11, bạn sửa “svm.SVR()” (dành cho bài toán hồi quy-regression) thành “svm.SVC” (dành cho bài toán phân lớp - classification); dòng 9 sửa “scoring=neg…” thành “scoring=f1” (dựa vào độ đo f1-score để chọn siêu tham số tốt nhất). Chạy và cho biết bộ siêu tham số nào tốt nhất, chạy phân lớp với bộ siêu tham số vừa chọn (khối lệnh [12] hình 4.32). Dùng classification_report để thể hiện kết quả trên tập huấn luyện và tập kiểm tra.

```python
from sklearn.datasets import load_digits
X, y = load_digits(return_X_y=True)
print(X.shape)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
	X, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import  classification_report
from sklearn.model_selection import GridSearchCV

tuned_parameters = [{'penalty': ['none'], 'solver': ['newton-cg', 'lbfgs', 'liblinear']}]

log_reg = GridSearchCV(
    	LogisticRegression(), tuned_parameters,
    	scoring='f1', cv=2
	)
log_reg.fit(X_train, y_train)
log_reg.best_params_
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(penalty='none', solver='lbfgs')
log_reg.fit(X_train, y_train)

from sklearn.metrics import classification_report

y_train_pred = log_reg.predict(X_train)
print('Train: classification report = ')
print(classification_report(y_train, y_train_pred))
y_test_pred = log_reg.predict(X_test)
print('Test: classification report = ')
print(classification_report(y_test, y_test_pred))
```

Câu 5: Như câu trên nhưng thực hiện với thuật toán LogisticRegression. Gợi ý: Bạn thế chỗ svm.SVR thành LogisticRegression(), thế chỗ dòng 1: “from sklearn.linear_model import LogisticRegression”, thế độ đo như câu 9, phần tham số các bạn đọc ở đây để biết tham số của LogisticRegression (gợi ý: solver,...). Dùng classification_report để thể hiện kết quả trên tập huấn luyện và tập kiểm tra.


```python
from sklearn.datasets import load_digits
X, y = load_digits(return_X_y=True)
print(X.shape)
```

```python
from sklearn.metrics import  classification_report
from sklearn import svm
from sklearn.model_selection import GridSearchCV

tuned_parameters = [{'kernel': ['rbf'], 'C': [1, 10, 100]},
                   {'kernel': ['linear'], 'C': [1, 10, 100]}]

svm_reg = GridSearchCV(
    	svm.SVC(), tuned_parameters,
    	scoring = 'f1', cv=2
	)
svm_reg.fit(X_train, y_train)
svm_reg.best_params_
```

```python
from sklearn import svm

svm_regr = svm.SVC(kernel='rbf', C=100)
svm_regr.fit(X_train, y_train)

from sklearn.metrics import mean_squared_error, mean_absolute_error
y_train_pred = svm_regr.predict(X_train)
print('Train: Mean squared error = ')
print(classification_report(y_train, y_train_pred))
print('Train: Mean absolute error = ')
print(classification_report(y_train, y_train_pred))

y_test_pred = svm_regr.predict(X_test)
print('Test: Mean squared error = ')
print(classification_report(y_test, y_test_pred))
print('Test: Mean absolute error = ')
print(classification_report(y_test, y_test_pred))
```
