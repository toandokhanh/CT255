#### 6 Từ dữ liệu dự đoán giá nhà boston house-prices, các bạn mô tả rõ các thuộc tính trong bộ này, mỗi thuộc tính để mô tả gì?



- CRIM tỷ lệ tội phạm trên đầu người theo thị trấn
- Tỷ lệ ZN của đất ở được quy hoạch cho các lô trên 25.000 bộ vuông.
- INDUS tỷ lệ mẫu đất kinh doanh phi bán lẻ trên mỗi thị trấn
-	Biến giả CHAS sông Charles (= 1 nếu đường giáp sông; 0 nếu không)
-	Nồng độ oxit nitric NOX (phần trên 10 triệu)
-	RM số phòng trung bình mỗi nhà ở
-	Tỷ lệ TUỔI của các căn hộ do chủ sở hữu sử dụng được xây dựng trước năm 1940.
-	Khoảng cách có trọng số của DIS đến năm trung tâm việc làm ở Boston.
-	Chỉ số RAD về khả năng tiếp cận các đường cao tốc xuyên tâm.
-	TAX thuế suất bất động sản toàn giá trị trên 10.000 đô la.
-	Tỷ lệ học sinh-giáo viên PT RATIO theo thị trấn.
-	B 1000(Bk - 0,63)^2 trong đó Bk là tỷ lệ người da đen theo thị trấn.
-	LSTAT % tình trạng thấp hơn của dân số.
-	MEDV Giá trị trung bình của những ngôi nhà do chủ sở hữu ở trong $1000.


####  7. Các bạn thay thế giá trị dự đoán trong boston house-price thành 1 trong 12 thuộc tính còn lại và chạy bài toán hồi quy với Rừng ngẫu nhiên với tỷ lệ tập huấn luyện là 50%, tập kiểm tra là 50%. Thực hiện lặp lại cho đến khi chạy xong hết 12 thuộc tính (trừ thuộc tính giá nhà, PRICE, đã trình bày trong giáo trình). Khối lệnh trình bày dự đoán của 1 thuộc tính trình bày trong 1 cell (có tổng cộng 12 cell cho câu này).




#### -CRIM




```python
import pandas as pd
from sklearn import datasets
boston = datasets.load_boston()
bos = pd.DataFrame(boston.data,columns = boston.feature_names)
bos['PRICE'] = boston.target
bos
```

```python
X = bos[['ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO',
         'B','LSTAT','PRICE']]
y=bos['CRIM']
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.5,random_state=1)

from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)


from sklearn.metrics import mean_squared_error, mean_absolute_error
y_train_pred = regr.predict(X_train)
print('Train: mean squared error = ',mean_squared_error(y_train,y_train_pred))
print('Train: Mean absolute error = ',mean_absolute_error(y_train, y_train_pred))

y_test_pred = regr.predict(X_test)
print('Test: mean squared error = ',mean_squared_error(y_test,y_test_pred))
print('Test: Mean absolute error = ',mean_absolute_error(y_test, y_test_pred))
```
