```python
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_olivetti_faces

def plot_images(images, total_images=20, rows=4, cols=5):
	fig = plt.figure()
	for i in range(total_images):
		img_grid = fig.add_subplot(rows, cols, i+1)
		img_grid. imshow(images[i])

faces = fetch_olivetti_faces()

images = faces.images

features = faces.data # featur:
targets = faces.target # targets
print(images.shape)
print(features.shape)
print(targets.shape)
# plot 10 images with 2 rows and 5 columns
plot_images(images, 20, 4, 5)
plt.show()
```

```python
from tensorflow.keras.preprocessing.image import load_img
def tim_face_images(images, image_cuaban, classid):
	j = 0 # gan id
	flag = 1
	fig = plt.figure(figsize=(24, 12)) # create a new figure window
	for i in range(len((targets))): # quet qua tung anh
		if faces.target[i] == classid:
			if j==0:
				img_grid = fig.add_subplot(2, 5, j+1)
                # plot features as image
				print('Anh cua ban')
				img_grid.imshow(image_cuaban)
			else:
				img_grid = fig.add_subplot(2, 5, j+1)
                # plot features as image
				print('classid',classid, 'sample_id',i)
				img_grid.imshow(images[i])
			j = j+1
			flag = 0
	if flag:
			print('Khong co khuon mat nao thoa classid ban da nhap vao')
```

```python
anh='avt1.png'
image_cuaban = load_img(anh, target_size=(64,64))
tim_face_images(images,image_cuaban,classid=5)
plt.show()
```
